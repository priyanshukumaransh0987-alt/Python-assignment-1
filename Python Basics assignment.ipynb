## 2. What is an interpreter in Python?
 An interpreter in Python is a program that executes Python code line by line, translating it into machine code that your computer can understand.

## 3. What are pre-defined keywords in Python?
 Keywords are special reserved words in Python that have predefined meanings and purposes.You cannot use them as variable names, function names, or identifiers.They tell Python how to interpret the code.
For example: if, else, while, for, def, class, etc.

## 4. Can keywords be used as variable names?
No, keywords cannot be used as variable names in Python.
Because keywords are reserved words that Python uses to define its syntax and rules. If you try to use them as variable names, Python will get confused and throw a SyntaxError.
## 5. What is mutability in Python?
  Mutability in Python means whether an object’s content value can be changed after it is created.

- Mutable objects → Can be changed in place without creating a new object.
- Immutable objects → Cannot be changed once created; any change makes a new object.
## 6. Why are lists mutable, but tuples are immutable?
### **List:**
- A list is designed to store a collection of items where changes are expected.

- You can add, remove, or update elements without creating a new list.
### **Tuple:**
- A tuple is meant to be a fixed collection of items (like constants).
- Once created, its contents cannot be changed.
## 7. What is the difference between “==” and “is” operators in Python?
In Python, the operator == checks whether two objects have the same value, while the operator is checks whether two variables point to the exact same object in memory. For example, two different lists with the same elements will be equal with == but not identical with is, because they occupy different memory locations. On the other hand, small integers and some strings may give True for both == and is, because Python internally caches them to save memory. In practice, you use == to compare values, and is when you specifically want to check identity, such as verifying if a variable is None.
## 8. What are logical operators in Python?
Logical operators in Python are used to combine conditional statements and return a Boolean value (True or False). The three logical operators are:
- and → returns True if both conditions are true.
- or → returns True if at least one condition is true.
- not → returns the opposite of the condition (negation).
## 9. What is type casting in Python?
Type casting in Python is the process of converting one data type into another. It allows you to change the type of a variable explicitly, such as converting a string to an integer, an integer to a float, or a list to a tuple. Python provides built-in functions for type casting, like int(), float(), str(), list(), tuple(), and set().
## 10. What is the difference between implicit and explicit type casting?
In Python, implicit type casting (also called type conversion) is done automatically by Python when it converts one data type to another without losing information. For example, when adding an integer and a float, Python automatically converts the integer into a float to avoid data loss. Explicit type casting, on the other hand, is when the programmer manually converts one data type into another using built-in functions like int(), float(), str(), list(), etc.
## 11. What is the purpose of conditional statements in Python?
The purpose of conditional statements in Python is to control the flow of a program by making decisions based on certain conditions. They allow the program to execute specific blocks of code only if a given condition is True, and optionally execute different code if the condition is False. This helps in building logic, handling different cases, and making programs dynamic.
## 12. How does the elif statement work?
The elif (short for else if) statement in Python is used to check multiple conditions one after another. It comes after an if statement and before an else statement. When the program runs, Python checks conditions in order:-
- If the first if condition is True, that block runs and the rest are skipped.
- If the first is False, Python moves to the elif condition.
- If an elif condition is True, its block runs and the rest are skipped.
- If none of the if or elif conditions are True, the else block (if present) runs.
## 13. What is the difference between for and while loops?
The main difference is that a for loop is used when you know in advance how many times you want to repeat a block of code (it iterates over a sequence like a list, string, or range), while a while loop is used when you want to keep executing a block of code until a condition becomes False.
- For loop: Iterates through items in a sequence or a range of numbers.
- While loop: Keeps running as long as the given condition is True.
## 14. Describe a scenario where a while loop is more suitable than a for loop?
A while loop is more suitable when the number of iterations is not known in advance and the loop needs to run until a certain condition is met. For example, if you are asking a user to enter a password until they type the correct one, you don’t know how many attempts they will take. In this case, a while loop is best because it continues running until the correct password is entered.

```python
# Question 10 -Write code to demonstrate type casting with list elements?
# Original list with string numbers
str_list = ["1", "2", "3", "4", "5"]

print("Original list:", str_list)

# Convert list elements from string to integer
int_list = [int(x) for x in str_list]
print("After converting to integers:", int_list)

# Convert list elements from string to float
float_list = [float(x) for x in str_list]
print("After converting to floats:", float_list)

# Convert list elements from string to boolean
bool_list = [bool(x) for x in str_list]
print("After converting to booleans:", bool_list)
```

```python
# Question 13-  Write a Python program to find the sum of all even numbers between 1 and 50.
# answer
# Initialize sum
total = 0

# Loop through numbers from 1 to 50
for i in range(1, 51):
    if i % 2 == 0:  # Check if even
        total += i

print("The sum of even numbers between 1 and 50 is:", total)
```

```python
# Question 14 - Write a program to reverse a string using a while loop.
# answe
# Take input from user
text = input("Enter a string: ")

# Initialize variables
reversed_text = ""
index = len(text) - 1

# Loop backwards using while
while index >= 0:
    reversed_text += text[index]
    index -= 1

print("Reversed string:", reversed_text)
```

```python
# QUESTION 15-  Write a Python program to calculate the factorial of a number provided by the user using a while loop.
# answer
# Take input from user
num = int(input("Enter a number: "))

# Initialize variables
factorial = 1
i = 1

# While loop to calculate factorial
while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is:", factorial)
```
